```python
import pandas as pd

df = pd.read_csv(r"C:\Users\saksh\Desktop\Retail-Sales-Forecasting\data\raw\walmart.csv")

df.head()
```

```python
import numpy as np

df['Simulated_Demand'] = df['Weekly_Sales'] + np.random.normal(0, 5000, len(df))

df[['Weekly_Sales', 'Simulated_Demand']].head()
```

```python
# Step 1: Date ko datetime me convert karo
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Step 2: Month extract karo
df['Month'] = df['Date'].dt.month

# Step 3: Seasonality apply karo
df['Seasonality'] = df['Month'].apply(lambda x: 1.2 if x in [11,12] else 0.9)

# Step 4: Seasonal sales calculate karo
df['Seasonal_Sales'] = df['Weekly_Sales'] * df['Seasonality']

# Output check
df[['Date','Month','Weekly_Sales','Seasonal_Sales']].head()
```

```python
# Initial stock define
initial_stock = 1000000

# Stock level calculate (cumulative sales minus)
df['Stock_Level'] = initial_stock - df['Weekly_Sales'].cumsum()

# Output check
df[['Date','Weekly_Sales','Stock_Level']].head()
```

```python
# Initial stock define
initial_stock = 20000000

# Stock level calculate (cumulative sales minus)
df['Stock_Level'] = initial_stock - df['Weekly_Sales'].cumsum()

# Output check
df[['Date','Weekly_Sales','Stock_Level']].head()
```

```python
# Stock-out condition
df['Stock_Out'] = df['Stock_Level'] < 0

# Check results
df[['Date','Stock_Level','Stock_Out']].head()
```

```python
# Reorder point (example logic)
reorder_point = df['Weekly_Sales'].mean() * 2

# Reorder decision
df['Reorder'] = df['Stock_Level'] < reorder_point

# Output check
df[['Date','Stock_Level','Reorder']].head()
```
